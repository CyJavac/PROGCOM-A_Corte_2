#Ecuaciones Empíricas

##1. Objetivo
Hallar la ecuación empírica del periodo de un anillo oscilante.



##2. Materiales


###2.1.Ecuación empírica
Es la ecuación que se obtiene de dos variables, compuestas por grupos de valores experimentales: $y = f (x)$ donde $x$ es la variable independiente, y $y$ es la variable dependiente.


###2.2. Calibrador
El calibre o calibrador es un instrumento de medición, principalmente de diámetros exteriores, interiores y profundidades, utilizado en el ámbito industrial.


###2.3. Anillos
Seis anillos de aluminio de distinto tamaño que utilizamos en el laboratorio para hallar su diámetro interno y externo en el que se basó el experimento.


###2.4. Metro
Instrumento para medir que consiste en una regla o en una cinta graduada que puede tener de 2 a 20 metros de longitud y que lleva marcada la división en decímetros, centímetros y milímetros.


###2.5. Sensor laser
Un sensor que con ayuda de un software puede detectar el movimiento y calcular el periodo de oscilación de un cuerpo.


## 3. Metodología
1. Se toman los anillos y los instrumentos de medici ́on.
2. Se toma la medida de los diámetros internos y externos de cada anillo.
3. Luego se pasaron los anillos por el sensor para obtener su periodo.
4. Se registran los datos obtenidos anteriormente en una tabla de excel.
5. Se halla el diámetro promedio de cada uno de los anillos.
6. Por último, graficamos los resultados, y, hallamos la ecuación empírica y el porcentaje de error.



##4. Análisis de Resultados

###4.2. Formulas utilizadas durante el experimento
Las formulas que usamos para este laboratorio fueron sencillas y las pudimos realizar con ayuda de excel. Estas fueron las formulas utilizadas:

* Ecuación empírica: $ y=f(x) $
* Diametro: D
* Pendiente: m
* Gravedad: $ g = 9.81m/s^2 $
* Promedio: $ \frac{\sum D_n}{n} $
* Periodo: T = $ 2\pi \cdot\sqrt{\frac{D}{g}} $
* Periodo²: $ T^2 $ = $ 4\pi^2 \cdot \frac{D}{g} $ = $\frac{4\pi^2}{g} \cdot D$
* Gravedad exprimental: $ \frac{4\pi^2}{m} $
* Porcentaje de error: $  |{\frac{g \cdot GravedadExperimental}{g}}| \cdot 100 $
