#Caída libre

##1. Objetivo

Comprobar mediante el experimento el valor de la gravedad y compararlo con el valor teorico $(9,81m/s^2)$.

##2. Introducción

La gravedad es una de las cuatro fuerzas fundamentales de la naturaleza (junto con el electromagnetismo, la fuerza nuclear fuerte y débil). 
Es una fuerza que atrae a los cuerpos entre sí, y depende de la distancia entre los cuerpos y de sus masas.
Los experimentos de Galileo lograron demostrar que los cuerpos en la tierra son atraídos hacia su centro con la misma aceleración, independientemente de su masa.
Esta aceleración rige las leyes de Newton, y a partir de estas leyes, se puede definir con gran precisión el valor de la gravedad mediante diferentes tipos de experimentos, por ejemplo: péndulo simple, plano de tiro oblicuo, caída libre, etc.

##3. Materiales


###3.1 Electroimán
Es un imán que tiene un conductor eléctrico, el cual crea a su alrededor un campo magnético cuando circula la corriente a través de él.



###3.2 Sensor laser

El sensor laser va a detectar cuando la rejilla interrumpa su trayectoria y tomara los datos de el tiempo en que sucedió y el espacio.



###3.3 Interruptor
Activador del mecanismo que hace que el electroimán suelte la rejilla.




###3.4 Rejilla
Rejilla utilizada como objeto de caída libre, con el cual el sensor puede detectar con facilidad la velocidad cada $0,01m$.
