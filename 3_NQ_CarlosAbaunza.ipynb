#Movimiento unidimensional

##1. Objetivo

Observar como actua un objeto con velocidad/aceleración constante según la distancia de movimiento.

##2. Introducción

El movimiento rectilíneo uniforme (MRU), es aquel con velocidad constante y cuya trayectoria es una línea recta. Cuando afirmamos que la velocidad es constante estamos afirmando que no cambia ni su valor ni la dirección del movimiento.

Un cuerpo realiza un movimiento rectilíneo uniformemente acelerado (MRUA) cuando su trayectoria es una línea recta y su aceleración es constante. Esto implica que la velocidad aumenta o disminuye su módulo de manera uniforme.

##3. Materiales

###3.1 Electroimán

Es un imán que tiene un conductor eléctrico, el cual crea a su alrededor un campo magnético cuando circula la corriente a través de él.

###3.2 Sensores laser

Los sensores laser van a detectar la velocidad y tiempo que toma el carro en llegar de un punto a otro.

###3.3 Polea

Polea utilizada para ejercer una fuerza sobre el carro y este se mueva a través de un sensor a otro.

###3.4 Carro

Objeto que será el que pase a través de los sensores gracias al peso que se ejerce al otro lado de la polea.

###3.5 Bomba de aire

Máquina que distribuirá aire por el mecanismo, eliminando la fricción que actua sobre el carro y asi obtener resultados mas precisos.

##4. Metodología

1. Cada grupo realizara tres lanzamientos del carro con una distancia diferente, aumentando cada vez $0,05m$ entre cada grupo.

2. Promediamos los resultados del tiempo de los tres lanzamientos, y los registramos el una tabla en excel.

3. Calculamos la velocidad de cada recorrido dividiendo la distancia en el tiempo.

4. Volvemos a realizar el mismo procedimiento, pero esta vez sin la polea y solo con la fuerza que ejerce el aire.

5. Con la ayuda de excel se realizan las graficas de: distancia vs tiempo, velocidad vs tiempo, y aceleración vs tiempo, de cada tipo de movimiento.

##5. Análisis de resultados

###5.1 Resultados experimentales

* Distancia ($m$).
* Aceleración ($m/s^2$)
* Tiempo ($s$).
* Velocidad ($m/s$)

###5.2 Fórmulas utilizadas durante el experimento

* Ecuación de la recta: $y=mx+b$
* Ecuación de la parábola: $y=ax^2+bx+c$
* Promedio: $\frac{\sum_{i=1}^{n} x_i}{n}$
* Velocidad: $\frac{Distancia}{Tiempo}$

##6. Conclusiones

* En el MRU la velocidad depende de la distancia y el tiempo.
* El MRU implica que la velocidad es constante, por lo tanto la aceleración será igual a cero.
