#Movimiento semiparabólico

##1. Objetivo

Hallar la velocidad inicial con la que cae un objeto formando un movimiento semiparabólico.

##2. Introducción

Este movimiento se caracteriza porque el cuerpo simultáneamente se mueve en el eje $x$ y en el eje $y$.

* En el eje $x$, se mueve durante todo el recorrido con velocidad constante es decir con un (MRU).
* El eje $y$, se mueve con un movimiento de caída de cuerpos; donde la velocidad inicial es cero.
* El tiempo que tarda el cuerpo en realizar la caída en $y$ es igual al tiempo que tarda el cuerpo en realizar el recorrido horizontal en $x$; porque, el cuerpo simultáneamente realiza los dos recorridos.

##3. Materiales

###3.1 Metro

Instrumento para medir que consiste en una regla o en una cinta graduada que puede tener de 2 a 20 metros de longitud y que lleva marcada la división en decímetros, centímetros y milímetros. 

###3.2 Esfera metálica

Esfera utilizada como objeto de caída.

###3.3 Papel carbón

Papel utilizado para marcar la ubicación de impacto de la esfera.

###3.4 Cuaderno

Cuaderno utilizado para amortiguar la caída de la esfera. También será donde se colocara el papel carbón y marcar el lugar de impacto de la esfera.

##4. Metodología

1. Se utilizó un cuaderno de 2,5cm de grosor para amortiguar la caída de la esfera y marcar el lugar de la caída.

2. Se realizan 3 lanzamientos, se miden las distancias a las que llega la bola en el eje $x$ con ayuda del metro, y se promedian los resultados.

3. Se vuelve a realizar el procedimiento con otras diez alturas diferentes.

4. Se restan los 2,5cm de grosor del cuaderno a la altura incial de cada lanzamiento.

5. Calculamos la velocidad inical en $x$ con la fórmula $\sqrt{\frac{1}{2}\cdot g\cdot \frac{x^2}{y}}$.

##5. Análisis de resultados

Los datos se tomaron en $cm$, y luego se pasaron a $m$ para coincidir con el sistema internacional de unidades.

###5.1 Resultados experimentales

* Distancia ($m$).
* Tiempo ($s$).
* Velocidad ($m/s$)

###5.2 Fórmulas utilizadas durante el experimento

* Promedio: $\frac{\sum_{i=1}^{n} x_i}{n}$
* X: $V_ix\cdot t$
* Tiempo: $t=\frac{x}{V_ix}$
* Y: $\frac{1}{2}\cdot g\cdot t^2$
* $V_ix$: $\sqrt{\frac{1}{2}\cdot g\cdot \frac{x^2}{y}}$

##6. Conclusiones

* Se pudo observar que un objeto con una velocidad inicial lanzado desde una rampa, dará un movimiento semiparabólico.
  
* El movimiento semiparabólico se presenta en dos dimensiones:   en   uno se   presenta   el   movimiento rectilíneo uniforme, perteneciente al eje $x$; y el otro movimiento es uniformemente acelerado presentado en el eje $y$, que es el mismo de caída libre.
